```python
import pandas as pd
import numpy as np
import sys, os
from string import ascii_lowercase
from glob import glob

CSV_DIR = "../csv/"
CSV_1_DIR = "../csv/sort_by_first_consonant/"
CSV_2_DIR = "../csv/sort_by_last_consonant/"
NPY_DIR = "../npy/"
JSON_DIR = "../json/"
```

<h3>List all data contents by iterating through each CSV file (Pandas dataframe)</h3>

```python
#list first 10 csv files from first consonant folder
for i in range(10):
    loop_path = f"{CSV_1_DIR}korean_{i} - Sheet1.csv"
    temp = pd.read_csv(loop_path)
    temp.dropna(how='all', axis=1, inplace=True)
    temp.dropna(how='all', axis=0, inplace=True)
    print(f"::: Table {i+1} / 10 ::: \n", temp, "\n")
```

```python
# list the rest of the csv files from the same folder
for i in range(65, 65+9):
    # 65 = A, 82 = R
    loop_path = f"{CSV_1_DIR}korean_{chr(i)} - Sheet1.csv"
    temp = pd.read_csv(loop_path)
    temp.dropna(how='all', axis=1, inplace=True)
    temp.dropna(how='all', axis=0, inplace=True)
    print(f"::: Table {chr(i)} :::\n", temp, "\n")
```

```python
# build 3D Numpy array, and save to .npy file
first_sheets = glob("../csv/sort_by_first_consonant/*.csv")
tmp_df = [pd.read_csv(f) for f in first_sheets]
arr = np.asarray(tmp_df)
```

```python
with open(f"{NPY_DIR}korean.npy", "wb") as op:
    np.save(op, arr)
    op.close()
```

```python
# do the same thing with a .json object
import json
from json import JSONEncoder

class NumpyArrayEncoder(JSONEncoder):
    def default(self, obj):
        return obj.tolist() if isinstance(obj, np.ndarray) else JSONEncoder.default(self, obj)
    
numpyData = {"array": arr}
encodedNumpyData = json.dumps(numpyData, cls=NumpyArrayEncoder, indent=4)
print(encodedNumpyData)

with open(f"{JSON_DIR}korean.json", "w", encoding="utf8") as op:
    op.write(encodedNumpyData)
    op.close()
```
