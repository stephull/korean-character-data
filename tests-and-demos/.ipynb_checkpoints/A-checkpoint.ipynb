```python
import pandas as pd
import numpy as np
import sys, os
from string import ascii_lowercase

CSV_DIR = "../csv/"
NPY_DIR = "../npy/"
JSON_DIR = "../json/"
```

<h3>List all data contents by iterating through each CSV file (Pandas dataframe)</h3>

```python
#list first 10 csv files
for i in range(10):
    loop_path = f"{CSV_DIR}korean_{i}.csv"
    temp = pd.read_csv(loop_path)
    temp.dropna(how='all', axis=1, inplace=True)
    temp.dropna(how='all', axis=0, inplace=True)
    print(f"::: Table {i+1} / 10 ::: \n", temp, "\n")
```

```python
# list the rest of the csv files
for i in range(65, 83):
    # 65 = A, 82 = R
    loop_path = f"{CSV_DIR}korean_{chr(i)}.csv"
    temp = pd.read_csv(loop_path)
    temp.dropna(how='all', axis=1, inplace=True)
    temp.dropna(how='all', axis=0, inplace=True)
    print(f"::: Table {chr(i)} :::\n", temp, "\n")
```
